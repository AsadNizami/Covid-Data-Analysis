```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 80)
```

```python
def get_dataframe(name):
    
    file_name = f'Output/{name}.csv'
    return pd.read_csv(file_name)

filtered_df = get_dataframe('dataset_filtered')
filtered_district_df = get_dataframe('TamilNadu_district_wise')
```

```python
def age_district(state):
    age_dist = dict()
    df = filtered_df[filtered_df['Detected State'] == state]
    display(df['Detected District'].unique())
    age_deceased_dist = dict() 
    

    for district in df['Detected District'].unique():
        age_dist[district] = df[(df['Detected District'] == district) & df['Age Bracket'].notnull()].shape[0]
        
    for district in df['Detected District'].unique():
        age_deceased_dist[district] = df[(df['Detected District'] == district) & df['Age Bracket'].notnull() & (df['Current Status'] == 'Deceased')].shape[0]
        
    print('Total number of cases for each district')
    display(age_dist)
    print('Total number of fatalities for each district')
    display(age_deceased_dist)
    
age_district('Bihar')
```

```python
def plot_gender_distribution(file, state, district):
    df = get_dataframe(file)
    df = df[df['District'] == district]
#     display(df)
    age_distribution = [f'{i}-{i+9}' for i in range(1, 80, 10)]
    age_distribution.append('80 or more')
    
    def get_male_cases():
        df_male_cases = df.iloc[:, 26: 26+9]
#         display(df_male_cases)
        return [sum(df_male_cases[col]) for col in df_male_cases]
    
    def get_female_cases():
        df_female_cases = df.iloc[:, 17: 17+9]
#         display(df_female_cases)
        return [sum(df_female_cases[col]) for col in df_female_cases]
    
    def get_male_fatalities():
        df_male_fatalities = df.iloc[:, 53:]
#         display(df_male_fatalities)
        return [sum(df_male_fatalities[col]) for col in df_male_fatalities]
        
    def get_female_fatalities():
        df_female_fatalities = df.iloc[:, 44: 44+9]
#         display(df_female_fatalities)
        return [sum(df_female_fatalities[col]) for col in df_female_fatalities]
    
    def plot_cases():
        width = 0.2
        fig = plt.subplots(figsize=(12, 8))
        
        male_cases = get_male_cases()
        female_cases = get_female_cases()
        
        br1 = range(9)
        br2 = [x + width for x in br1]


        plt.bar(br1, male_cases, color ='b', width = width,
            edgecolor ='grey', label ='Male Cases')
        plt.bar(br2, female_cases, color ='r', width = width,
            edgecolor ='grey', label ='Female Cases')

        plt.title(f'Covid cases distribution age and gender wise for {district}, {state}')
        plt.xlabel('Age', fontweight ='bold', fontsize = 15)
        plt.ylabel('Covid Cases', fontweight ='bold', fontsize = 15)
        plt.xticks([r + width for r in range(9)],
            age_distribution)

        plt.legend()
        plt.show()
        
    def plot_fatalities():
        width = 0.2
        fig = plt.subplots(figsize=(12, 8))
        
        male_fatalities = get_male_fatalities()
        female_fatalities = get_female_fatalities()
        
        br1 = range(9)
        br2 = [x + width for x in br1]
        plt.bar(br1, male_fatalities, color ='b', width = width,
            edgecolor ='grey', label ='Male Fatalities')
        plt.bar(br2, female_fatalities, color ='r', width = width,
            edgecolor ='grey', label ='Female Fatalities')
        
        plt.title(f'Fatalities distribution age and gender wise for {district}, {state}')
        plt.xlabel('Age', fontweight ='bold', fontsize = 15)
        plt.ylabel('Covid Fatalities', fontweight ='bold', fontsize = 15)
        plt.xticks([r + width for r in range(9)],
            age_distribution)

        plt.legend()
        plt.show()
        
#     get_male_cases()
#     get_female_cases()
#     get_male_fatalities()
#     get_female_fatalities()
    plot_cases()
#     plot_fatalities()    
    
# plot_gender_distribution('Karnataka', 'Udupi')
```

```python
plot_gender_distribution('Karnataka_district_wise', 'Karnataka', 'Udupi')
plot_gender_distribution('Karnataka_district_wise', 'Karnataka', 'Kalaburagi')
plot_gender_distribution('TamilNadu_district_wise', 'Tamil Nadu', 'Chengalpattu')
plot_gender_distribution('TamilNadu_district_wise', 'Tamli Nadu', 'Thiruvallur')
plot_gender_distribution('Bihar_district_wise', 'Bihar', 'Patna')
plot_gender_distribution('Bihar_district_wise', 'Bihar', 'Bhagalpur')
```
